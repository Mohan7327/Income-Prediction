## **Importing Necessary Libararies**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## **Importing Data**

```python
data = pd.read_csv('Income data.csv')
data
```

```python
# rows with missing values represented as'?'.
data_1 = data[data.workclass == '?']
data_1
```

```python
# rows with missing values represented as'?'.
data = data[data.workclass != '?']
data
```

```python
data = data[data.occupation != '?']
data = data[data['native-country'] != '?']
```

```python
data['occupation'].value_counts()
```

## **Data Preparation**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_categorical = data.select_dtypes(include='object')
data_categorical
```

```python
le = LabelEncoder()
data_categorical = data_categorical.apply(le.fit_transform)
data_categorical
```

```python
data = data.drop(data_categorical.columns,axis = 1)
data = pd.concat([data,data_categorical],axis = 1)
data 
```

```python
data.info()
```

```python
# convert target variable income to categorical
data['income'] = data['income'].astype('category')
```

## **Model Building**

```python
from sklearn.model_selection import train_test_split
```

```python
x= data.drop('income',axis = 1)
y= data[['income']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 12)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## **Model Training**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion="gini", splitter="best", max_depth=4,random_state=12)
model.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(16,10))
plot_tree(model,filled = True, rounded = True)
plt.show()
```

## **Model Testing**

```python
y_train_pred = model.predict(x_train) # For Training Data
y_train_pred
```

```python
y_test_pred = model.predict(x_test) # For Testing Data
y_test_pred
```

## **Model Evaluation**

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
accuracy_score(y_train,y_train_pred) # For Training Data
```

```python
accuracy_score(y_test,y_test_pred) # For Testing Data
```

```python
confusion_matrix(y_test,y_test_pred)
```

#**Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV,KFold,cross_val_score
```

```python
GridSearch_CV = GridSearchCV(estimator = model, param_grid = {'max_depth':[4,5,6,7,8],'criterion':['gini','entropy']},cv=5, n_jobs=-1)
GridSearch_CV.fit(x_train,y_train)
```

```python
GridSearch_CV.best_params_
```

```python
GridSearch_CV.best_score_
```

```python
cv_score = cross_val_score(estimator=model, X=x, y=y,cv=5)
print('5 Fold CV Scores   :', cv_score)
print('Mean Accuracy      :', round(cv_score.mean(),4))
print('Standard Deviation :', round(cv_score.std(),4))
```

```python
DT_model = DecisionTreeClassifier(criterion="gini", splitter="best", max_depth=8,random_state=12)
DT_model.fit(x_train,y_train)
```

```python
y_test_pred = DT_model.predict(x_test) # For Testing Data
y_test_pred
```

```python
accuracy_score(y_test,y_test_pred) # For Testing Data
```
